## plot log of Raleigh p-value on symlog scale with inverted y axis and minor tick marks


```python
import matplotlib.pyplot as plt
import numpy as np
from matplotlib.ticker import (MultipleLocator, FormatStrFormatter, AutoMinorLocator)
import matplotlib.patheffects as pe
from matplotlib import rcParams
rcParams['font.sans-serif']=["arial"]
```

```python
def plot_logp(x_values, y_log10, ytickvalues, ylimvalues, addcolor, xlabels):
    fig, ax = plt.subplots(figsize=(2,1.9))
    plt.scatter(x_values, y_log10, color=addcolor, linewidth=0.5, s=70, zorder=1) #s=100 for scatter
    plt.plot(x_values, y_log10, zorder=0, color=addcolor, lw=3, path_effects=[pe.Stroke(linewidth=2.5, foreground='black'), pe.Normal()]) #s=100 for scatter    
    plt.xticks(x_values, xlabels)
    plt.yticks(ytickvalues)
    ax.spines['right'].set_visible(False)
    ax.spines['top'].set_visible(False)
    plt.xlabel("ES Frequency (Hz)", size=12)
    plt.ylabel('-log10 (p-value)', size=12)
    ax.set_ylim(ylimvalues)
    plt.setp(ax.spines.values(), linewidth=2)
    plt.xticks(fontsize=11)
    plt.yticks(fontsize=11)
    minorLocator = MultipleLocator(10)
    ax.tick_params(length=5, width=2)
    sigline=-np.log10(0.05)
    plt.axhline(y=sigline, color='#FF00FF', linestyle='-')
    plt.show()
```

### For mouse V1 Cells

```python
#For V1 Pyr
x_values_200=[0, 1, 2, 3]
xlabels = ["0", "8", "30", "140"]
y_values_200=[5.0e-01, 1.35E-40, 1.23E-24, 4.69E-38]
y_log10_200= -(np.log10(y_values_200))
ytickvalues_200 = np.arange(0, 46, 15)
ylimvalues = (-2.5, 49)
addcolor = "#009b81"

plot_logp(x_values_200, y_log10_200, ytickvalues_200, ylimvalues, addcolor, xlabels)
```

```python
#For V1 Pvalb
x_values=[0, 1, 2, 3]
xlabels = ["0", "8", "30", "140"]
#the Rayleigh p-value at 200 nA is a super-tiny #, so the "1e-145" value here is just for plotting purposes
#this is indicated in the figure as a >150 with the two lines showing that it extends further
y_values_200=[8.5e-01, 0.043, 0.000022, 1e-145]  
y_log10_200= -(np.log10(y_values_200))
ytickvalues_200 = np.arange(0, 151, 50)
ylimvalues = (-8, 165)
addcolor = "#cd3449"

plot_logp(x_values_200, y_log10_200, ytickvalues_200, ylimvalues, addcolor, xlabels)
```

```python
#For V1 SST
x_values=[0, 1, 2, 3]
xlabels = ["0", "8", "30", "140"]
y_values_200=[5.4e-01, 0.0036, 3e-31, 1.44e-133]  
y_log10_200= -(np.log10(y_values_200))
ytickvalues_200 = np.arange(0, 140, 30)
ylimvalues = (-8, 140)
addcolor = "#fe7f11"

plot_logp(x_values, y_log10_200, ytickvalues_200, ylimvalues, addcolor, xlabels)
```
