```python
import matplotlib.pyplot as plt
import numpy as np
from matplotlib import rcParams
rcParams['font.sans-serif']=["arial"]
```

```python
def plot_sd(x_val, y_val, coloring, ylims, savefigpath=None):
    fig, ax = plt.subplots(figsize=(2,1.9))
    ax.scatter(x_val, y_val, color=coloring, linewidth=0.5, s=70, zorder=1)
    ax.plot(x_val,y_val, color=coloring, lw=3)
    ax.spines['right'].set_visible(False)
    ax.spines['top'].set_visible(False)
    ax.set_xlabel('', fontsize=18, weight='medium', labelpad=15)
    ax.set_ylabel('', fontsize=18, weight='medium', labelpad=15)
    # ax.set_yscale('log')
    ax.set_ylim(ylims)
    plt.setp(ax.spines.values(), linewidth=2)
    ax.tick_params(length=5, width=2)
    # ax.tick_params(axis = 'both', which = 'major', labelsize = 12)
    if savefigpath:
        plt.savefig(savefigpath, bbox_inches='tight')
    return ax
```

```python
x_values=['con', '50', '100', '200']

#Pyr 8 Hz
y_values8pyr=[145.75, 138.58, 109.29, 90.80]
#Pyr 30 Hz
y_values30pyr=[156.04, 117.22, 124.50, 97.69]
#Pyr 140 Hz
y_values140pyr=[150.18, 127.99, 91.40, 84.03]

#PValb 8 Hz
y_values8pv=[216.69, 166.24, 165.02, 163.77]
#PValb 30 Hz
y_values30pv=[182.04, 173.54, 166.57, 153.12]
#PValb 140 Hz
y_values140pv=[171.78, 122.08, 122.91, 84.83]

#SST 8 Hz
y_values8sst = [169.53, 162.64, 154.32, 145.20]
#SST 30 Hz
y_values30sst = [180.95, 153.43, 131.45, 112.92]
#SST 140 Hz
y_values140sst = [143.16, 109.66, 97.33, 87.33]
```

```python
#Pyr
ylims = (70,170)
pyrvals=(y_values8pyr, y_values30pyr, y_values140pyr)
celltype = "Pyr"

i=0
for j in [8,30,140]:
    savefilename = "AngleSDplot_200nA_" + str(j) + "Hz" + celltype + ".svg"
    plot_sd(x_values, pyrvals[i], "#009b81", ylims)      
    #savefigpath = savefigfolder + savefilename
    #plt.savefig(savefigpath, bbox_inches='tight')
    plt.show()
    i=i+1
```

```python
#Pvalb
ylims = (70,230)
pv_vals=(y_values8pv, y_values30pv, y_values140pv)
celltype = "PValb"
    
i=0
for j in [8,30,140]:
    savefilename = "AngleSDplot_200nA_" + str(j) + "Hz" + celltype + ".svg"
    plot_sd(x_values, pv_vals[i], "#cd3449", ylims)      
    #savefigpath = savefigfolder + savefilename
    #plt.savefig(savefigpath, bbox_inches='tight')
    plt.show()
    i=i+1
```

```python
#SST
ylims = (70,230)
sstvals=(y_values8sst, y_values30sst, y_values140sst)
celltype = "SST"

i=0
for j in [8,30,140]:
    savefilename = "AngleSDplot_200nA_" + str(j) + "Hz" + celltype + ".svg"
    plot_sd(x_values, sstvals[i], "#fe7f11", ylims)      
    #savefigpath = savefigfolder + savefilename
    #plt.savefig(savefigpath, bbox_inches='tight')
    plt.show()
    i=i+1
```

# Levene's test

```python
from scipy import stats
from scipy.stats import levene
import random
import numpy as np
```

1) in the pandas dataset, make arrays of the spike_phases, and then perform Levene's test from scipy. 
The code template is here, you just need to make a lot of arrays for comparisons!!!

```python
#Opening pickled dictionary
import time
import pickle
start_time = time.time()

pictable_V1Pyr_pd = "../data/pickletable_V1Pyr_pd"
pictable_V1Pvalb_pd = "../data/pickletable_V1Pvalb_pd"
pictable_V1SST_pd = "../data/pickletable_V1SST_pd"

#open the file for reading
open_V1Pyr_obj_pd = open(pictable_V1Pyr_pd, 'rb')
open_V1Pvalb_obj_pd= open(pictable_V1Pvalb_pd, 'rb')
open_V1SST_obj_pd = open(pictable_V1SST_pd, 'rb')

#load the object from the file into new_table_PC
all_PCs = pickle.load(open_V1Pyr_obj_pd)
all_Pvalbs = pickle.load(open_V1Pvalb_obj_pd)
all_SSTs= pickle.load(open_V1SST_obj_pd)

print("Done")
print("This code took", "%.2f"%(time.time() - start_time), "seconds to run.")
```

```python
# when there is ES (ES is non-zero)
supra_PCs = all_PCs[(all_PCs['num_spikes_A']>0) & (all_PCs['ex_amp_nA']!=0) & (all_PCs['ex_dur_ms']==13000) & (all_PCs['ex_el_distance_mu']==50)]
supra_Pvalbs = all_Pvalbs[(all_Pvalbs['num_spikes_A']>0) & (all_Pvalbs['ex_amp_nA']!=0) & (all_Pvalbs['ex_dur_ms']==13000) & (all_Pvalbs['ex_el_distance_mu']==50)]
supra_SSTs = all_SSTs[(all_SSTs['num_spikes_A']>0) & (all_SSTs['ex_amp_nA']!=0) & (all_SSTs['ex_dur_ms']==13000) & (all_SSTs['ex_el_distance_mu']==50)]
```

```python
#control data (no ES, i.e. ES is zero)
control_supra_PCs = all_PCs[(all_PCs['num_spikes_A']>0) & (all_PCs['ex_amp_nA']==0) & (all_PCs['ex_dur_ms']==13000) & (all_PCs['ex_el_distance_mu']==50)]
control_supra_Pvalbs = all_Pvalbs[(all_Pvalbs['num_spikes_A']>0) & (all_Pvalbs['ex_amp_nA']==0) & (all_Pvalbs['ex_dur_ms']==13000) & (all_Pvalbs['ex_el_distance_mu']==50)]
control_supra_SSTs = all_SSTs[(all_SSTs['num_spikes_A']>0) & (all_SSTs['ex_amp_nA']==0) & (all_SSTs['ex_dur_ms']==13000) & (all_SSTs['ex_el_distance_mu']==50)]
```

```python
#For the ES experiments
#You will have need to make nine arrays, for each cell class

#stimamp, stimfreq: (50, 8), (50, 30), (50, 140)
#stimamp, stimfreq: (100, 8), (100, 30), (100, 140)
#stimamp, stimfreq: (200, 8), (200, 30), (200, 140)
#for supra_PCs, for supra_Pvalbs, for supra_SSTs

stimfreq = 140
stimamp = 50
firstgroup_amp_freq = supra_PCs[(supra_PCs['ex_amp_nA']==stimamp) & (supra_PCs['ex_frequency']==stimfreq)]
justphases  = firstgroup_amp_freq['spike_phase_A_corrected']
justphases_onelist  = justphases.sum()

#name the array accordingly (e.g. PCs_8_140, etc)
PCs_140_50 = np.array(justphases_onelist)
```

```python
#For the CONTROL experiments
#You will have need to make three arrays, for each cell class

#stimamp, stimfreq: (0, 8), (0, 30), (0, 140)
#for supra_PCs, supra_Pvalbs, supra_SSTs
#remember to take sample of control spikes instead of duplicating them

stimfreqcon = 140
stimampcon = 0
n_sample = 1311
secondtgroup_amp_freq = control_supra_PCs[(control_supra_PCs['ex_amp_nA']==stimampcon) & (control_supra_PCs['ex_frequency']==stimfreqcon)]
justphases_con  = secondtgroup_amp_freq['spike_phase_A_corrected']
justphases_con_onelist  = justphases_con.sum()
justphases_con_onelist_sampled = random.sample(justphases_con_onelist, n_sample)

#name the array accordingly (e.g. PCs_8_140, etc)
PCs_CON140_50_random = np.array(justphases_con_onelist_sampled)
```

```python
#just a quick test
levstatef, levpvalef = levene(PCs_140_50, PCs_CON140_50_random)
print levpvalef
```
